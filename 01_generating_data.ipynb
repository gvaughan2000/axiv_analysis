```python
# Module Imports
import dask.dataframe as dd
from dask.distributed import Client, LocalCluster

# Sets up a dask cluter to run on local machine
cluster = LocalCluster() 

# Sets the maximum number of workers 
cluster.adapt(minimum=1, maximum=8) 

# Creates a client to enable interaction with the cluster
client = Client(cluster)

# Explicitly defining data types
defined_dtypes = {'id': str, 'title': str, 'comments':  str, 'journal-ref':  str, 'id':  str}
# reads in data
df = dd.read_json('arxiv.json', blocksize="256 MiB", dtype=defined_dtypes, )
df.head(5)
```

```python
defined_dtypes = {'id': str, 'title': str, 'comments':  str, 'journal-ref':  str, 'id':  str}

df = dd.read_json('arxiv.json', blocksize="256 MiB", dtype=defined_dtypes, )
df.head(5)
```


## Removing papers without comments

```python
df = df[['id', 'title', 'comments', 'journal-ref', 'categories']]
```

```python
df = df[df['comments']!='None']
df.head(5)
```

```python
# Filter to remove records containing 'COVID' in the title 
df = df.loc[~df["title"].str.contains("covid", case=False)]
```

```python
# Some titles included new line characters
df['title'] = df['title'].str.replace('\n', '').replace('"', '')
```

```python
final_df = df.compute()
```

```python
final_df.to_csv("without_covid.csv", index=False)
```
